## PIE CHART

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = 'Trips.csv'
data = pd.read_csv(file_path)
data['Per Person Price'] = data['Per Person Price'].astype(float)
agg_data = data.groupby('Destination')['Per Person Price'].sum().reset_index()
top_n = 10
top_destinations = agg_data.nlargest(top_n, 'Per Person Price')
other_destinations = agg_data[~agg_data['Destination'].isin(top_destinations['Destination'])]
others_row = pd.DataFrame({'Destination': ['Others'], 'Per Person Price': [other_destinations['Per Person Price'].sum()]})
others_row = pd.DataFrame(others_row)
agg_data_grouped = pd.concat([top_destinations, others_row])
colors = sns.color_palette('husl', n_colors=len(agg_data_grouped))
plt.figure(figsize=(10, 8))
plt.pie(
    agg_data_grouped['Per Person Price'],
    labels=None,  
    autopct=None,  
    startangle=90,
    colors=colors
)
plt.title(f'Top {top_n} Tourism Per Person Price')
legend_labels = [f'{destination}: {color}' for destination, color in zip(agg_data_grouped['Destination'], colors)]
plt.legend(legend_labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

## LINE GRAPH

```python
import matplotlib.pyplot as plt
data_sorted = data.sort_values(by='Package Name')
selected_data = data_sorted.iloc[::len(data_sorted)//10, :]
x_axis = selected_data['Places Covered']
y_axis = selected_data['Package Name']
plt.figure(figsize=(12, 6))
plt.plot(x_axis, y_axis, color='black', linestyle='-.', marker='o', markerfacecolor='r')
plt.xlabel('Destination')
plt.ylabel('Package Name')
plt.title( 'Destinations across Package Name')
plt.xticks(rotation=45, ha='right', fontsize=10)  
plt.tight_layout()  
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
grouped_data = data.groupby('Package Name')['Per Person Price'].mean().reset_index()
selected_data = grouped_data.iloc[::len(grouped_data)//10, :]
plt.plot(selected_data['Package Name'], selected_data['Per Person Price'], marker='o', linestyle='-', color='c')
plt.ylabel('Average Per Person Price')
plt.xlabel('Package Name')
plt.title('Average Per Person Price Across Package Names')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## SCATTER PLOT

```python
import matplotlib.pyplot as plt
Travel_Date = data['Travel Date']
Airline = data['Airline']
selected_travel_date_values = Travel_Date.unique()[::len(Travel_Date.unique())//10]
plt.scatter(Travel_Date, Airline, color='m', marker='o')
plt.xlabel('Travel Date')
plt.ylabel('Airline')
plt.title('Scatter Plot of Airline Over Travel Date')
plt.xticks(selected_travel_date_values, rotation=45, ha='right')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```

## BAR GRAPH
