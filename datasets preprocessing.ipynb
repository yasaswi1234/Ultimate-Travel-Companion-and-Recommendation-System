```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("trips.csv")
```

```python
data
```

```python
data
```

```python
data.fillna('0',inplace=True)
```

```python
data
```

```python
data
```

```python
data = data.drop(columns=['Page Url'])
```

```python
data
```

```python
# Define the allowed package types
allowed_types = ['Deluxe', 'Standard', 'Premium', 'Luxury', 'Budget']

# Filter the DataFrame to keep only the rows with allowed package types
data = data[data['Package Type'].isin(allowed_types)]
```

```python
data
```

```python
# Define the allowed package types
allowed_types = ['Makemytrip']

# Filter the DataFrame to keep only the rows with allowed package types
data = data[data['Company'].isin(allowed_types)]
```

```python
data
```

```python
print(data.columns)
```

```python
data['Travel Date'] = pd.to_datetime(data['Travel Date'], format='%d-%m-%Y', errors='coerce')

# Filter rows where the conversion was successful
data = data[~data['Travel Date'].isnull()]
```

```python
data
```

```python
data = data[data['Hotel Details'].apply(lambda x: isinstance(x, str))]
```

```python
data
```

```python
data.to_csv('pp_trips.csv', index=False)
```
