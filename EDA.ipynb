```python
plt.figure(figsize=(10, 8))
sns.histplot(df["Per Person Price"])  
plt.title("Distribution of Per Person Price")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Per Person Price'], bins=30, color='skyblue', alpha=0.7)
plt.xlabel('Per Person Price')
plt.ylabel('Frequency')
plt.title('Histogram of Per Person Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Per Person Price', data=df, color='lightgreen')
plt.xlabel('Per Person Price')
plt.title('Boxplot of Per Person Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Per Person Price'], df['Price Per Two Persons'], color='salmon', alpha=0.5)
plt.xlabel('Per Person Price')
plt.ylabel('Price Per Two Persons')
plt.title('Scatter Plot between Per Person Price and Price Per Two Persons')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
